```python
 def panel_score(data,indrug):
    """
    used in readinData function
    
    Arugments:
    data -- wordvecs
    indrug -- drugname
    
    Returns:
    scores -- list of cosine similarity scores
    """
    scores=[]
    for i in ['haloperidol','risperidone','fluoxetine','mirtazapine','nortriptyline','olanzapine']:
        scores.append(np.dot(data[indrug],data[i])/ ( np.linalg.norm(data[indrug])*np.linalg.norm(data[i])))
    return scores
```

```python
#load the vector into dataframe 
def loadframe(filename):
    """
    used in readinData function
    
    Arugments:
    filename -- aer2vec vectors
    
    Returns:
    df -- a dataframe contains the aer2vec vectors
    """
    wvectors=svu.readfile(filename)
    df=pd.DataFrame(np.array(wvectors[1][:]))
    cleanterms=[]
    for term in iwvectors[0]:
        cleanterms.append(term[0:])
        
    df['name']=cleanterms
    return df
```

```python
def readinData(dataset,chem_prop,indicator):
    """
    readin the data
    
    Arugments:
    dataset -- wordvecs
    indicator -- had options "panel", "normalize"
    chem_prop -- the reference sets, either Martin or Gao which contains a list of drugs as well as the
                 chemical properties. 
    
    Returns:
    dfc -- a dataframe has both vectors and the reference set of choice
    """
    df1 = loadframe(dataset)
    wordvecs = df1.set_index('name').T.to_dict('list')
    if indicator=='panel':
        result= list()
        mean=list()
        for i in wordvecs:
            result.append(panel_score(wordvecs,i))
            mean.append(np.mean(panel_score(wordvecs,i)))
            n_array_scores = np.array(result)
            n_array_mean=np.array(mean)
            
        n_array_mean_r= np.reshape(n_array_mean, (len(wordvecs), 1)) 
        np_array_b= np.array(df1)
#the first 6 columns are the score and the 7th column is the mean, the rest are vectors
        output= np.concatenate((n_array_scores, n_array_mean_r, np_array_b), axis=1)
        df=pd.DataFrame(output)
        df.rename(columns={df.columns[-1]: "name", df.columns[0]:"haloperidol_normalized", df.columns[1]:"risperidone_normalized", df.columns[2]:"fluoxetine_normalized",
                          df.columns[3]:"mirtazapine_normalized", df.columns[4]:"nortriptyline_normalized", df.columns[5]:"olanzapine_normalized", df.columns[6]:"mean_6_normalized"}, inplace = True)
       
    elif indicator=='normalize':
        for i in wordvecs:
            wordvecs[i] = wordvecs[i] / np.linalg.norm(wordvecs[i])
        df=pd.DataFrame(wordvecs, columns=list(df1.columns.values)).T
        df.index.name = 'name'
        df.reset_index(inplace=True)

    else: 
        df=pd.DataFrame(df1, columns=list(df1.columns.values)).T
        
        
    dfc=pd.merge(df, chem_prop,on='name', how='inner')
    return dfc
```

```python
def setTestTrain(df_train, dataset1, df_test, dataset2):
    """
    Arugments:
    df_train, dataset1, df_test, dataset2 -- df_train, and df_test are the dataframes
    dataset1 -- "Gao"(training set)
    dataset2 -- "Martin"(testing set)
    
    Returns:
    df_h_train-- a dataframe (training)
    df_h_test-- a dataframe (testing)
    """
    df_h_train=pd.DataFrame.merge(df_train,df_c, on=['name'],how='inner')
    df_h_train['ID']= str(dataset1)
    df_h_test=pd.DataFrame.merge(df_test,df_c, on=['name'],how='inner')
    df_h_test['ID']= str(dataset2)
    return df_h_train, df_h_test
```

```python
def z_score(df):
    """
    calculate Z scores to certain columns in the dataframe but not all
    
    Arugments:
    df -- a dataframe that contains the columns of interest
    
    Returns:
    df -- only z transformed the designated columns but not all
    """
    df[['MW','xlogP','mean_Tanimoto','haloperidol_tanimoto','risperidone_tanimoto','fluoxetine_tanimoto','mirtazapine_tanimoto','nortriptyline_tanimoto','olanzapine_tanimoto','mean_Tanimoto']]=zscore(df[['MW','xlogP','mean_Tanimoto','haloperidol_tanimoto','risperidone_tanimoto','fluoxetine_tanimoto','mirtazapine_tanimoto','nortriptyline_tanimoto','olanzapine_tanimoto','mean_Tanimoto']])
    return df
```

```python
def pca(df, df2):
    """
    PCA

    Arugments:
    df--df_all, a dataframe that has all the original variables
    df2--df_zscore_all, the z score transformed data
    
    Returns:
    X_z-- a dataset has the MW and xlogP both Z transformed and the 6 prinicple components
    
    """
    X_i=df.loc[:,~df.columns.isin(['p_np','cid2','cid','name','name_y','name_x','pass','MW','xlogP','hal_zscore_D','hal_zscore_T','ris_zscore_D','ris_zscore_T','flu_zscore_D','flu_zscore_T','mir_zscore_D','mir_zscore_T','nor_zscore_D','nor_zscore_T','ola_zscore_D','ola_zscore_T','Z_score_mean_T','Z_score_mean_D',
                                   'z_scoreus','double_transform_D','double_transform_T','mean_z_us','hal_z_transformed','ris_z_transformed','flu_z_transformed','mir_z_transformed','nor_z_transformed','ola_z_transformed','raw_t_score_z','index','pc_1', 'pc_2','pc_3','pc_4','pc_5','pc_6','cid_x','cid_y',
                                    'haloperidol_normalized','risperidone_normalized','fluoxetine_normalized','mirtazapine_normalized','nortriptyline_normalized','olanzapine_normalized','mean_6_normalized','haloperidol_tanimoto','risperidone_tanimoto','fluoxetine_tanimoto','mirtazapine_tanimoto','nortriptyline_tanimoto','olanzapine_tanimoto','mean_Tanimoto',
                                    'name','p_np','cid2','cid','name_y','name_x','pass','hal_zscore_D','hal_zscore_T','ris_zscore_D','ris_zscore_T','flu_zscore_D','flu_zscore_T','mir_zscore_D','mir_zscore_T','nor_zscore_D','nor_zscore_T','ola_zscore_D','ola_zscore_T','Z_score_mean_T','Z_score_mean_D','z_scoreus','double_transform_D','double_transform_T','mean_z_us',
                                  'hal_z_transformed','ris_z_transformed','flu_z_transformed','mir_z_transformed','nor_z_transformed','ola_z_transformed','haloperidol_normalized','risperidone_normalized','fluoxetine_normalized','mirtazapine_normalized','nortriptyline_normalized','olanzapine_normalized','mean_6_normalized','raw_t_score_z',
                                    'index','pc_1', 'pc_2','pc_3','pc_4','pc_5','pc_6','haloperidol_tanimoto','risperidone_tanimoto','fluoxetine_tanimoto','mirtazapine_tanimoto','nortriptyline_tanimoto','olanzapine_tanimoto','mean_Tanimoto','haloperidol_tanimoto_zscore','risperidone_tanimoto_zscore','fluoxetine_tanimoto_zscore','mirtazapine_tanimoto_zscore','nortriptyline_tanimoto_zscore','olanzapine_tanimoto_zscore',
                                    'mean_Tanimoto_Z_transformed','MW_log','ID'])]
    df2.reset_index(inplace=True) 
    pca = PCA(n_components=6)
    x = X_i.values
    principalComponents = pca.fit_transform(x)
    principalDf = pd.DataFrame(data = principalComponents
             , columns = ['pc_1', 'pc_2','pc_3',
                          'pc_4','pc_5','pc_6'])
    df=principalDf.astype('float')
    df_f_1= pd.concat((df2, df),axis=1)
    
    X_z= df_f_1.loc[:,df_f_1.columns.isin(['MW','xlogP','pc_1', 'pc_2','pc_3','pc_4','pc_5','pc_6'])]
    return X_z
```

```python
def data_prep(df):
    """
    Input the dataframe and output the dataframe for modeling
    
    Arugments:
    df -- a dataframe that contains all the column for modeling
    
    Returns:
    X_s, X_b, X_e, X_g, X_d, output the dataframes for modeling
    """
       
    #molecular weight
    X_s=df.loc[:,df.columns.isin(['MW','xlogP'])]
    
    X_b=df.loc[:,df.columns.isin(['MW','xlogP','haloperidol_normalized','risperidone_normalized','fluoxetine_normalized','mirtazapine_normalized','nortriptyline_normalized','olanzapine_normalized'])]
    #basic features plus 500d
    X_e=df.loc[:,~df.columns.isin(['name','p_np','cid2','cid','name_y','name_x','pass','hal_zscore_D','hal_zscore_T','ris_zscore_D','ris_zscore_T','flu_zscore_D','flu_zscore_T','mir_zscore_D','mir_zscore_T','nor_zscore_D','nor_zscore_T','ola_zscore_D','ola_zscore_T','Z_score_mean_T','Z_score_mean_D','z_scoreus','double_transform_D','double_transform_T','mean_z_us',
                                  'hal_z_transformed','ris_z_transformed','flu_z_transformed','mir_z_transformed','nor_z_transformed','ola_z_transformed','haloperidol_normalized','risperidone_normalized','fluoxetine_normalized','mirtazapine_normalized','nortriptyline_normalized','olanzapine_normalized','mean_6_normalized','raw_t_score_z',
                                    'index','pc_1', 'pc_2','pc_3','pc_4','pc_5','pc_6','haloperidol_tanimoto','risperidone_tanimoto','fluoxetine_tanimoto','mirtazapine_tanimoto','nortriptyline_tanimoto','olanzapine_tanimoto','mean_Tanimoto','haloperidol_tanimoto_zscore','risperidone_tanimoto_zscore','fluoxetine_tanimoto_zscore','mirtazapine_tanimoto_zscore','nortriptyline_tanimoto_zscore','olanzapine_tanimoto_zscore',
                                    'mean_Tanimoto_Z_transformed','level_0','ID'])]
    
    X_g=df.loc[:,df.columns.isin(['MW','xlogP','mean_Tanimoto'])]
    X_d=df.loc[:,df.columns.isin(['MW','xlogP','haloperidol_tanimoto','risperidone_tanimoto','fluoxetine_tanimoto','mirtazapine_tanimoto','nortriptyline_tanimoto','olanzapine_tanimoto'])]

    return X_s, X_b, X_e, X_g, X_d
```

```python
def model(X_train, y_train, X_test, y_test):
#l2 regularized model    
    model=LogisticRegression(penalty='l2',random_state=0, solver='liblinear',max_iter=10000)
    model.fit(X_train, y_train)
    result=model.score(X_test, y_test)
    preds= model.predict(X_test)
    predprob=model.predict_proba(X_test)[:,1]
    true=y_test
    
    return roc_auc_score(true, predprob)
```

```python
def skfrocauc(X_train, y_train, X_test, y_test):
    tprs = []
    mean_fpr = np.linspace(0,1,100)
    
    model = LogisticRegression(penalty='l2',solver='liblinear',max_iter=10000)
    model.fit(X_train, y_train)
    probas_ = model.predict_proba(X_test)[:,1]
    fpr, tpr, thresholds = roc_curve(y_test, probas_, pos_label=1)

    roc_auc = auc(fpr, tpr)
    return  tpr, fpr, roc_auc
```
