```python
#6 panel score, cosine similarity 
def panel_score(data,indrug):
    """
    used in readinData function
    
    Arugments:
    data -- wordvecs
    indrug -- drugname
    
    Returns:
    scores -- list of cosine similarity scores
    """
    scores=[]
    for i in ['haloperidol','risperidone','fluoxetine','mirtazapine','nortriptyline','olanzapine']:
        scores.append(np.dot(data[indrug],data[i])/ ( np.linalg.norm(data[indrug])*np.linalg.norm(data[i])))
    return scores
```

```python
#load the vector into dataframe 
def loadframe(filename):
     """
    used in readinData function
    
    Arugments:
    filename -- aer2vec vectors
    
    Returns:
    df -- a dataframe contains the aer2vec vectors
    """
    wvectors=svu.readfile(filename)
    df=pd.DataFrame(np.array(wvectors[1][:]))
    cleanterms=[]
    for term in iwvectors[0]:
        cleanterms.append(term[0:])
        
    df['name']=cleanterms
    return df
```

```python
#read in the data
def readinData(dataset,indicator,chem_prop):
    """
    readin the data
    
    Arugments:
    dataset -- wordvecs
    indicator -- had options "panel", "normalize"
    chem_prop -- the reference sets, either Martin or Gao which contains a list of drugs as well as the
                 chemical properties. 
    
    Returns:
    dfc -- a dataframe has both vectors and the reference set of choice
    """
    df1 = loadframe(dataset)
    wordvecs = df1.set_index('name').T.to_dict('list')
    if indicator=='panel':
        result= list()
        mean=list()
        for i in wordvecs:
            result.append(panel_score(wordvecs,i))
            mean.append(np.mean(panel_score(wordvecs,i)))
            n_array_scores = np.array(result)
            n_array_mean=np.array(mean)
            
        n_array_mean_r= np.reshape(n_array_mean, (len(wordvecs), 1)) 
        np_array_b= np.array(df1)
#the first 6 columns are the score and the 7th column is the mean, the rest are vectors
        output= np.concatenate((n_array_scores, n_array_mean_r, np_array_b), axis=1)
        df=pd.DataFrame(output)
        df.rename(columns={df.columns[-1]: "name", df.columns[0]:"haloperidol_normalized", df.columns[1]:"risperidone_normalized", df.columns[2]:"fluoxetine_normalized",
                          df.columns[3]:"mirtazapine_normalized", df.columns[4]:"nortriptyline_normalized", df.columns[5]:"olanzapine_normalized", df.columns[6]:"mean_6_normalized"}, inplace = True)

    else: 
        df=pd.DataFrame(df1, columns=list(df1.columns.values)).T
        
        
    dfc=pd.merge(df, chem_prop,on='name', how='inner')
    return dfc
```

```python
def z_score(df):
    """
    calculate Z scores to certain columns in the dataframe but not all
    
    Arugments:
    df -- a dataframe that contains the columns of interest
    
    Returns:
    df -- only z transformed the designated columns but not all
    """
    df[['MW','xlogP','mean_Tanimoto','haloperidol_tanimoto','risperidone_tanimoto','fluoxetine_tanimoto','mirtazapine_tanimoto','nortriptyline_tanimoto','olanzapine_tanimoto','mean_Tanimoto']]=zscore(df[['MW','xlogP','mean_Tanimoto','haloperidol_tanimoto','risperidone_tanimoto','fluoxetine_tanimoto','mirtazapine_tanimoto','nortriptyline_tanimoto','olanzapine_tanimoto','mean_Tanimoto']])
    return df
```

```python
#calculate the auc, also the confidence interval 
def skfrocauc(vecs, labels):
    """
    calculate roc auc and calcuate TPR, FPR, upper, lower, std 
    
    Arugments:
    vecs -- an array of features and vectors 
    labels -- an array of label

    Returns:
    mean_tpr, mean_tprs, mean_fpr, mean_aucs, tprs_upper, tprs_lower, aucs_std -- all the variables for the ROC- AUC 
    curve. 
    
    """
    aucs = []
    tprs = []
    mean_fpr = np.linspace(0,1,100)
    all_aucs=list()
    all_tprs=list()
    
#repeat 20 times
    for i in range(20):    
        for train,test in StratifiedKFold(n_splits=5, shuffle=True).split(vecs, labels):
            model = LogisticRegression(penalty='l2',solver='liblinear',max_iter=10000)
            model.fit(vecs[train], labels[train])
            probas_ = model.predict_proba(vecs[test])[:,1]
            fpr, tpr, thresholds = roc_curve(labels[test], probas_, pos_label=1)
            tprs.append(np.interp(mean_fpr, fpr, tpr))
            tprs[-1][0] = 0.0
            roc_auc = auc(fpr, tpr)
            aucs += [auc(fpr, tpr)]
    
            mean_tpr = np.mean(tprs, axis=0)
            mean_tpr[-1] = 1.0
            mean_auc = auc(mean_fpr, mean_tpr)
    
        all_tprs.append(mean_tpr)
        all_aucs.append(mean_auc)
    
    mean_tprs= np.mean(all_tprs, axis=0)
    mean_aucs= np.mean(all_aucs, axis=0)
    aucs_std= np.std(all_aucs)
    std_tprs= np.std(all_tprs, axis=0)
    tprs_upper = np.minimum(mean_tprs + 1.96*std_tprs, 1)
    tprs_lower = np.maximum(mean_tprs - 1.96*std_tprs, 0)
    return  mean_tpr, mean_tprs, mean_fpr, mean_aucs, tprs_upper, tprs_lower, aucs_std
```

```python
def mean_confidence_interval(data, confidence=0.95):
    """
    calculate the confidence interval for the L1-regularized model
    
    Arugments:
    data-- a list of auc 
    confidence -- confidence interval 
    
    Returns:
    m, m-h, m+h -- m, mean; m-h lower, m+h upper 
    
    """
    a = 1.0 * np.array(data)
    n = len(a)
    m, se = np.mean(a), stats.sem(a)
    h = se * stats.t.ppf((1 + confidence) / 2., n-1)
    return m,m-h,m+h
```

```python
def calculateL1(X_z_zscore, X_d_zscore, X_g_zscore, X_s_zscore, X_b_zscore, X_e_zscore):
    """
    calculate l1, calling the skftest_l1(df)
    
    Arugments:
    X_z_zscore, X_d_zscore, X_g_zscore, X_s_zscore, X_b_zscore, X_e_zscore -- 
    the dataframes that were prepared in the addition_prep function
    
    """
    fivecvFeaturesOnly_Z=list()
    fiveFeaMeanTanimotoZ=list()
    fiveFea6SimilarityZ=list()
    fivecvSim_Z_6_weDerived=list()
    fivecvFeature_Z_6PCA=list()
    fivecvFeatures_Z_vectors=list()

    for i in range(20):
        fivecvFeaturesOnly_Z.append(skftest_l1(X_s_zscore)[1])
        
        fiveFeaMeanTanimotoZ.append(skftest_l1(X_d_zscore)[1])

        fiveFea6SimilarityZ.append(skftest_l1(X_g_zscore)[1])
 
        fivecvSim_Z_6_weDerived.append(skftest_l1(X_b_zscore)[1])
    
        fivecvFeature_Z_6PCA.append(skftest_l1(X_z_zscore)[1])

        fivecvFeatures_Z_vectors.append(skftest_l1(X_e_zscore)[1])
    
    
    print('cv5_L1_FeaturesOnly_Z/'+str(mean_confidence_interval(fivecvFeaturesOnly_Z)))
    print('cv5_L1_MW_xlogP_Mean_Tanimoto_Z_score_1/'+ str(mean_confidence_interval(fiveFeaMeanTanimotoZ)))
    print('cv5_L1_MW_xlogP_6_Tanimoto_simiarity_Z_score/'+ str(mean_confidence_interval(fiveFea6SimilarityZ)))
    print('cv5_L1_MW_xlogP_Z_transformed_6_panel_score/'+str(mean_confidence_interval(fivecvSim_Z_6_weDerived)))
    print('cv5_L1_MW_xlogP_Z_transformed_6_PCA/' + str(mean_confidence_interval(fivecvFeature_Z_6PCA)))
    print('cv5_L1_Features_Vectors_Z_transformed/'+str(mean_confidence_interval(fivecvFeatures_Z_vectors)))
```
